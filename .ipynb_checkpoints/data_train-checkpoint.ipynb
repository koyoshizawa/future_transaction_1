```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.layers.recurrent import LSTM
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
# csv読み込み
file_name = 'data_list.csv'
df = pd.read_csv(file_name, header=None)
df.head()
```

```python
# 学習データと教師データに分ける
def make_dataset(df):
    
    data , target = [], []
    col_len = len(df.columns)
    
    for i, row in df.iterrows():
        data.append(row[0:col_len - 2])
        target.append(row[col_len - 1])
    
    re_data = np.array(data).reshape(len(data), col_len - 2, 1)
    re_target = np.array(target).reshape(len(data), 1)
    
    return re_data, re_target

#     ランダムフォレスト用
#     return data, target

# x -> 学習データ y -> 教師データ
x, y = make_dataset(df)
```

```python
# 検証データと訓練データに分割
N_train = int(len(x) * 0.9)
N_validation = len(x) - N_train

x_train, x_validation, y_train, y_validation = train_test_split(x, y, test_size=N_validation)
y_train = [int(y) for y in y_train]
y_validation = [int(y) for y in y_validation]

# 0,1を[1, 0] [0, 1]にする
# num_classes = 2
# y_train = to_categorical(y_train, num_classes)
# y_validation = to_categorical(y_validation, num_classes)
```

```python
#ランダムフォレスト (変数が時系列の場合は適してなさそう　結果も良くなかった)
from sklearn.metrics import (roc_curve, auc, accuracy_score)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf = clf.fit(x_train, y_train)
pred = clf.predict(x_validation)
fpr, tpr, thresholds = roc_curve(y_validation, pred, pos_label=1)
auc(fpr, tpr)
accuracy_score(pred, y_validation)
```

```python
# 変数別に重要度可視化
import matplotlib.pyplot as plt
%matplotlib inline

features = [str(i) for i in range(19)]
importances = clf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(6,6))
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.show()
```

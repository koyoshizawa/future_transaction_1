```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.layers.recurrent import LSTM
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
# csv読み込み
file_name = 'data_list.csv'
df = pd.read_csv(file_name, header=None)
df.head()
```

```python
# 学習データと教師データに分ける
def make_dataset(df):
    
    data , target = [], []
    col_len = len(df.columns)
    
    for i, row in df.iterrows():
        data.append(row[0:col_len - 2])
        target.append(row[col_len - 1])
    
    re_data = np.array(data).reshape(len(data), col_len - 2, 1)
    re_target = np.array(target).reshape(len(data), 1)
    
    return re_data, re_target

# g -> 学習データ h -> 教師データ
g, h = make_dataset(df)
```

```python
# 検証データと訓練データに分割
N_train = int(len(g) * 0.9)
N_validation = len(g) - N_train

x_train, x_validation, y_train, y_validation = train_test_split(g, h, test_size=N_validation)
y_train = [int(y) for y in y_train]
y_validation = [int(y) for y in y_validation]

num_classes = 2
y_train = to_categorical(y_train, num_classes)
y_validation = to_categorical(y_validation, num_classes)
```
